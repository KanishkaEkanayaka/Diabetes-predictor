```python
from keras.models import Sequential
from keras.layers import Dense
import numpy
import pandas as pd
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv("diabetes.csv")
df.head()
```

```python
model=Sequential()  # create NN layers sequentially
```

```python
model.add(Dense(12,input_dim=8,activation='relu'))
model.add(Dense(20,activation='relu')) # hidden layer
model.add(Dense(20,activation='relu')) # hidden layer
model.add(Dense(22,activation='relu')) # hidden layer
model.add(Dense(1,activation='sigmoid')) # output layer
```

Calculating loss in neural networks is crucial for training and evaluating their performance. It quantifies how well the model's predictions deviate from the actual values. Different algorithms are suited for various tasks, depending on the problem type (regression vs. classification) and the nature of the outputs. Here are some common loss functions and their use cases:

Regression Problems:

Mean Squared Error (MSE): Squares the differences between predicted and true values, then averages them. Suitable for continuous outputs where larger errors are penalized more heavily. Easy to interpret and optimize.

Use case: Predicting housing prices, stock market values, or sensor readings.
Mean Absolute Error (MAE): Takes the absolute differences between predicted and true values, then averages them. Less sensitive to outliers compared to MSE but doesn't penalize large errors as much.

Use case: Robust regression tasks where outliers are present, like weather forecasting or demand forecasting.
Huber Loss: Combines the strengths of MSE and MAE. Smoothly transitions between them, making it less sensitive to outliers than MSE while still penalizing large errors.

Use case: When dealing with both outliers and the need to penalize larger errors, such as in robust regression tasks.
Classification Problems:

Binary Cross-Entropy: Measures the difference between the predicted probability of a class and the actual label (0 or 1). Suitable for binary classification problems.

Use case: Image classification (cat vs. dog), sentiment analysis (positive vs. negative), fraud detection (fraudulent vs. legitimate).
Categorical Cross-Entropy: Extends binary cross-entropy to multi-class problems. Measures the difference between the predicted probability distribution over all classes and the one-hot encoded true label.

Use case: Multi-class image classification (dog, cat, bird), classifying handwritten digits (0-9).
Hinge Loss: Used in support vector machines (SVMs), measures the margin between the correct class score and the scores of other classes. Encourages a larger margin between classes for better separability.

Use case: Text classification, image classification with well-separated classes.
Remember, choosing the right loss function depends on your specific problem and dataset characteristics. Experimenting with different options can help you achieve better model performance.

In addition to these, there are other specialized loss functions designed for specific tasks, such as ranking problems or structured output prediction. It's always a good practice to consult relevant resources and documentations for the specific framework or library you're using, as they might offer additional loss function options tailored to your needs.

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
inputs = df.drop(['Outcome'],axis='columns')
```

```python
target=df.Outcome # dependant attribute
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(inputs,target,test_size=0.2)
```

```python
model.fit(X_train,y_train,epochs=100,batch_size=10)
```

```python
scores=model.evaluate(X_test,y_test)
```

```python
print(model.metrics_names[1],scores[1]*100)
```
